```python
import nlp
import vis
```

```python
with open("data/session notes.txt", 'r') as f:
    test_text = f.read()

triples = nlp.logic(test_text)
graph = vis.logic(triples)
```

```python
import networkx as nx
pos = nx.nx_pydot.graphviz_layout(graph, prog="neato")
```

```python
edges = [edge for edge in graph.edges];
nx.draw_networkx_edge_labels(
    graph,
    pos,
    edge_labels={
        (u, v): graph.get_edge_data(u, v, _)["label"]
        for u, v, _ in [edge for edge in graph.edges]
    },
);
```

```python
from IPython.display import Image, display

def view_pydot(pdot):
    plt = Image(pdot.create_png())
    display(plt)
```

```python
pydotgraph = nx.drawing.nx_pydot.to_pydot(graph)

view_pydot(pydotgraph)
```
